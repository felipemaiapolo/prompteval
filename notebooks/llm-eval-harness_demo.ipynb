We cluster templates and selected those close to the centroids. This step guarantees some diversification in templates.

```python
#Clustering
model = SentenceTransformer(EMBEDDER) 
X = model.encode([t[-1] for t in templates])

seed=0
while True:
    kmeans = KMeans(n_clusters=N_TEMPLATES, random_state=seed, n_init="auto").fit(X)
    selected_templates = pairwise_distances(kmeans.cluster_centers_, X, metric='euclidean').argmin(axis=1)
    if np.unique(selected_templates).shape[0]==N_TEMPLATES:
        break
    else:
        seed+=1
        
selected_templates = np.sort(selected_templates)
templates = [templates[i] for i in selected_templates]
templates[:5]
```

## Evaluating

Evaluating an LLM on MMLU subjects using `N_TEMPLATES` prompt templates. We start sampling:

```python
seen_examples_dict = {}
examples_dict = {}

for topic in TOPICS:
    budget = 4*min(N_TEMPLATES,N_TOPICS[topic])
    seen_examples_dict[topic] = StratSample(np.zeros((N_TEMPLATES, N_TOPICS[topic])).astype(bool), budget, random_seed=0)
    examples_dict[topic] = {}
    for i,d in enumerate(seen_examples_dict[topic]):
        examples_dict[topic][f"mmlu_multi_prompt_{topic}_{i}"] = [j for j,x in enumerate(d) if x] 
```

```python
examples_dict['astronomy']['mmlu_multi_prompt_astronomy_0'] #this template will eval these examples
```

Running `lm-eval-harness`:

```python
%%time
results = {}
for topic in tqdm(TOPICS):
    results[topic] = simple_evaluate(
        model="hf",
        model_args=f"pretrained={MODEL_DIR}",
        tasks=f"mmlu_multi_prompt_{topic}",
        batch_size=4,
        num_fewshot=5,
        examples=examples_dict[topic],
        log_samples=True,
        device="cuda"
    )
```

Processing the data to a format we can apply PromptEval:

```python
Ys = {}
for topic in TOPICS:
    Ys[topic] = -np.ones((N_TEMPLATES, N_TOPICS[topic]))
    for i in range(N_TEMPLATES):
        temp = list(results[topic]['samples'].values())[i]
        for d in temp:
            Ys[topic][i,d['doc_id']] = d['acc']
```
